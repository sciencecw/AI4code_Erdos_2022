## Features

1. Auto-refresh - updating cells updates the TOC.
2. Cell Jumping - Clicking TOC links jump to the cell (yep it's a TOC)
3. Cell Nesting - H1 > H6
4. Toggle-able - Able to close (clicking button) and open (by rerunning the cell).
5. Minimal overhead

## Source
If interested in how it works, review the following:

```python
%%javascript
(function () {
    const doc = IPython.page.site_div_element.context;
    // Add some space.
    const site = doc.querySelector("#site");
    site.style.paddingRight = "300px";
    // Toc already exists.
    if (doc.querySelector("#toc")) return;
    // Helpers.
    const $ = ({parent=doc.body, type="div", id, cssText, value=""}) => {
        let el = doc.createElement(type);
        if (id) el.id = id;
        if (cssText) el.style.cssText = cssText;
        el.innerHTML = value;
        parent.appendChild(el);
        return el;
    };
    const aStyle = h => {
        const indent = (parseInt(h.nodeName.replace("H", "")) - 1) * 8;
        return `font-weight:bold; display:block; padding-left:${indent}px; overflow:hidden; white-space:nowrap; text-overflow:ellipsis;`;
    };
    // Initialize TOC.
    $({
        id: "toc",
        cssText: "position:fixed; top:0; right:0; width:300px; height:100%; border-left:1px solid #ccc; padding:20px; overflow-y:auto;",
    });
    $({
        id: "toc-close",
        value: "Close",
        type: "button",
        cssText: "position: fixed; top: 0; right: 0;" 
    }).addEventListener("click", () => {
        site.style.paddingRight = "0px";
        doc.body.removeChild(doc.querySelector("#toc"));
    }, false);
    
    let key = "";
    const buildTree = () => {
        const toc = doc.querySelector("#toc");
        if (!toc) return;
        const headers = [...site.querySelectorAll("h1, h2, h3, h4, h5, h6")];
        // Prevent unnecessary renders.
        const newKey = headers.map(h => h.nodeName + h.id).join(",");
        if (newKey != key) {
            key = newKey;
            toc.innerHTML = headers.map(h => `<a href="#${h.id}" style="${aStyle(h)}">${h.innerText.trim()}</a>`).join("");
        }
        window.requestAnimationFrame(buildTree);
    };
    buildTree();
})();
```
