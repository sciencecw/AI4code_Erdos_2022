We'll create a model that predicts crop yields for apples and oranges (target variables) by looking at the average temperature, rainfall, and humidity (input variables or features) in a region. Here's the training data:

****

```python
import pandas as pd
import torch
import numpy as np
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]], dtype='float32')
```

```python
#yeild in tons
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119]], dtype='float32')
```

lets convert np array into pytorch tensor


```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print("inputs",inputs)
print("targets",targets)
```

```python
#weight and bias
w = torch.rand(2,3,requires_grad= True)
b= torch.rand(2, requires_grad= True)
print(w)
print(b)
```

```python
def model(x):
    
    return x @ w.t()+ b

#Our model is simply a function that performs a matrix multiplication of the 
#inputs and the weights w (transposed) and adds the bias b (replicated for each observation).
```

x * wT + b

```python
preds = model(inputs)
print(preds)
```

```python
#lets compare this result to actual targets
print(targets)
print("diffrence bitween actual and predicted ", preds-targets)
```

**Loss function**

#Calculate the difference between the two matrices (preds and targets).
#quare all elements of the difference matrix to remove negative values.
#Calculate the average of the elements in the resulting matrix.

```python
def lossf(t1,t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
loss =lossf(preds,targets)
print(loss)
```

**Compute gradients**

```python
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
w
w.grad
```

```python
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
```

```python
loss = lossf(preds, targets)
print(loss)
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```
