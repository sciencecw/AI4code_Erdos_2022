```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
data=pd.read_csv("/kaggle/input/dissolved-oxygen-prediction-in-river-water/train.csv")
data.head()
```

```python
data.info()
```

Missing values


```python
null_columns=list(data.columns[data.isna().sum()>100])
data.drop(null_columns,axis=1,inplace=True)
```

```python
null_columns=list(data.columns[data.isna().sum()>100])
data.drop(null_columns,axis=1,inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
print("Columns with missing values:",(data.isna().sum(axis=0)!=0).sum())
print("Rows with missing values:",(data.isna().sum(axis=1)!=0).sum())      
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.isna().sum().sum()
```

```python
data
```

Splitting Dataset 

```python
data.drop('Id',axis=1,inplace=True)
```

```python
data
```

```python
y=data['target']
X=data.drop('target',axis=1)
```

```python
print(X)
print(y)
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
pd.DataFrame(X)
```

**Training Regression**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
model=LinearRegression()
model.fit(X_train,y_train)
model_R2=model.score(X_test,y_test)
```

```python
print("Model R2 Score:",model_R2)
```

**New Problem :Predicting High or Low Dissolving oxygen**

```python
y.mean()
```

```python
y_new=pd.qcut(y,q=2,labels=[0,1])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y_new,train_size=0.7)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
model_acc=model.score(X_test,y_test)
```

```python
print("Model Accuracy:",model_acc)
```
