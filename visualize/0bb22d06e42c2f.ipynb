### CO2 Emissions - Data Visualisation

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt 
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
# Load File
csv = '../input/co2-emissions-1960-2018/CO2_Emissions_1960-2018.csv'
df = pd.read_csv(csv,index_col='Country Name')
```

```python
df.head()
```

```python
print('data dimensions:', df.shape)
```

```python
years = list(map(str, range(1960, 2018)))

years
```

```python
df['Total'] = df.sum(axis=1)
```

```python
df.isnull().sum()
```

```python
df.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df.head()

# transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.plot(kind='area', 
             alpha=0.25,  # 0 - 1, default value alpha = 0.5
             stacked=False,
             figsize=(20, 10))

plt.title('CO2 emissions (metric tons per capita) of Top 5 contributors')
plt.ylabel('CO2 emissions (metric tons)')
plt.xlabel('Years')

plt.show()
```

```python
df.sort_values(['Total'], ascending=True, axis=0, inplace=True)

# get the top 5 entries
df_least_5 = df.head()

# transpose the dataframe
df_least_5 = df_least_5[years].transpose()

df_least_5.head()
```

```python

df_least_5 = df.tail(5)
df_least_5 = df_least_5[years].transpose()
df_least_5.index = df_least_5.index.map(int)
ax = df_least_5.plot(kind='area', alpha=0.35, figsize=(20, 10))
df_least_5.plot(kind = 'area', alpha=0.35, figsize=(20, 10))
ax.set_title('CO2 emissions (metric tons per capita) of Least contributoring countries')
ax.set_ylabel('CO2 emissions(metric tons)')
ax.set_xlabel('Years')
```
