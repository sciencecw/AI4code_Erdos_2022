```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
df_data.T
```

```python
df_data.describe().T
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_data.corr(), annot=True, cmap='Spectral')
plt.show()
a = df_data.hist(figsize=(20, 20))
```

We observe there are outliers in the Glucose, BloodPressure, BMI, Insulin, DiabetesPedigreeFunction

For Glucose, BloodPressure, BMI, Insulin, they can't be zero, we will replace them with the average value of the corresponding attribute.

```python
data = df_data.copy()
for attr in ['Glucose', 'BloodPressure', 'Insulin', 'BMI']:
    data[attr] = data[attr].replace(0, data[attr].mean())
a = data.hist(figsize=(20, 20))
```

```python
X = data.to_numpy()[:,:-1]
Y = data.to_numpy()[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix

# fit the model
for kernel in ('linear', 'sigmoid', 'rbf'):
    clf = svm.SVC(kernel=kernel)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f'{kernel}\'s accuracy: {np.mean(y_pred == y_test)}')
    # conf_mat = confusion_matrix(y_test, y_pred)
    # sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='g')
    # plt.show()

    # print(conf_mat)
    # print(classification_report(y_test, y_pred))
    # print(y_pred)
    # print(y_test)

clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
Y_pred = clf.predict(X)
print(f'We choose linear kernel, and run on the original data. Accuracy: {np.mean(Y_pred == Y)}')
conf_mat = confusion_matrix(Y_pred, Y)
sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='g')
plt.show()
```

```python
from sklearn.neighbors._nearest_centroid import NearestCentroid
clf = NearestCentroid()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f'KNN\'s accuracy: {np.mean(y_pred == y_test)}')
conf_mat = confusion_matrix(y_pred, y_test)
sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='g')
plt.show()
```
