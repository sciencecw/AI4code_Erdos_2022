```python
import pandas as pd
import altair as alt
```

> First, we're going to read a file, **business-licences.csv**, from our dataset and save it as an object named business.

```python
business = pd.read_csv('../input/city-of-vancouver-business-licence-statistics/business-licences.csv', delimiter=';')
business.head()
```

> When we read in the file above, it returns an error in the red line - DtypeWarning: Columns (13) - because one of the columns in the data frame uses mixed data types as an element in their values.

> However, it is okay to ignore it for now because we are going to sort out the data by selecting specific columns we want to use. 😉

![](https://cdn.sivanaspirit.com/wp-content/uploads/2018/04/03121706/eab31da4d00a3ac017600f89d96f0e5d_featured.jpg)

> It is always good to look over the data types of our data frame before starting.

```python
business.dtypes
```

> Now, we're going to sort out the data by selecting specific columns we want to use.

> We're going to sort out the values in a column named 'BusinessType' and drop off some columns we don't want to use and save it as an object named business_sorted.

> It makes our data frame looks clean and neat. 🧬 ✂️

```python
business_sorted = business.sort_values(by='BusinessType').drop(columns=['LicenceRSN','LicenceNumber','LicenceRevisionNumber','Street','Unit','UnitType','House','City','Province','Country','PostalCode','ExtractDate','Geom'])
business_sorted.head()
```

> Next, to get the latest information applicable, we're going to filter a column named 'FOLDERYEAR' to 21 and 'Status' to 'Issued' and save it as an object named business_filtered.

```python
business_filtered = business_sorted[(business_sorted['FOLDERYEAR'] == 21) & (business_sorted['Status'] == 'Issued')]
business_filtered.head()
```

> Perfect! Now we are at the right place to play with this sorted data. 👍

![](https://media.makeameme.org/created/good-job-way.jpg)

> Once we look up, we want to know what 'Business Name' or 'Business Trade Name' has the largest number of employees committed to the local job market. 💰

> This time, let's sort out the values in a column named 'NumberofEmployees' in descending order and show the top 10 business names that have hired the largest number of employees and save it as an object named business_top_employees.
# Vancouver's Top 10 Business Names That Have Hired The Largest Number of Employees.

```python
business_top_employees = business_filtered.sort_values(by='NumberofEmployees', ascending=False)
business_top_employees.head(10)
```

> Interesting!

> As a result, we can see that PNE(Pacific National Exhibition) has the greatest employment number followed by Calderon's Painting Ltd. and Source Han International Enterprise Inc.

> By the way, we want to know how many business types have existed in Vancouver.

> With a data frame named 'business_filtered', we're going to group it by a column named 'BusinessType' and save it as an object named 'business_type_grouped'.

```python
business_type_grouped = business_filtered.groupby(by='BusinessType')
business_type_grouped
```

> Here we have 172 different types of business in Vancouver, according to the city of Vancouver's dataframe.

```python
business_type_grouped.size()
```

> Awesome!

> Let's find out the total number of employees in each business type and save it as an object named 'business_type'.

# Vancouver's Top 10 Business Types That Have Hired The Largest Number of Employees.

```python
business_type = business_type_grouped.sum().drop(columns=['FOLDERYEAR','FeePaid']).sort_values(by='NumberofEmployees', ascending=False).reset_index()
business_type.head(15)
```

> Then, let's make it to a chart! 📊

```python
business_type_chart = alt.Chart(business_type.head(15), width=700, height=400).mark_bar().encode(
        x=alt.X('BusinessType:N', sort='y', title='Business Type'),
        y=alt.Y('NumberofEmployees:Q', title='Number of Employees')
).properties(title="Vancouver's Top 15 Business Types Have The Largest Number of Employees")

business_type_chart
```

# Vancouver's Top 10 Financial Institution That Have Hired The Largest Number of Employees.



```python
financial_institution_grouped = business_filtered[business_filtered['BusinessType'] == 'Financial Institution'].groupby(by='BusinessName')
financial_institution_grouped
```

```python
financial_institution = financial_institution_grouped.sum().drop(columns=['FOLDERYEAR','FeePaid']).sort_values(by='NumberofEmployees', ascending=False).reset_index()
financial_institution.head(15)
```

> Interesting!

> In the city of Vancouver, Central 1 Credit Union has the largest number of employees. It seems like the community-based banking institutions are taking a pretty big spot in the local financial sector.

![](https://www.central1.com/wp-content/uploads/2021/09/Central-1-Vancity-First-West-press-release-newsfeed.png)

> Then, what are the local areas the most bankers are working?

```python
fi_location_grouped = business_filtered[business_filtered['BusinessType'] == 'Financial Institution'].groupby(by='LocalArea')
fi_location_grouped
```

```python
fi_location = fi_location_grouped.sum().drop(columns=['FOLDERYEAR','FeePaid']).sort_values(by='NumberofEmployees', ascending=False).reset_index()
fi_location.head(10)
```
