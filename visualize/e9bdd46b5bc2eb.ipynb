**The data was available in json format, and moderately clean data, the objective of this notebook is to give a gist of how to change the json format to dataframe format... The attempt was made on the data of India artist**

```python
from pathlib import Path
FILE_NAME = ['artists_ids_IN','artist_ids','data_artists_IN','data_artists','music']
data_dir = Path('/kaggle/input/spotify-artist-data-for-some-artists/')

# the path name and all files names have been stored here, so that loading the data will be quicker
```

```python
for name in FILE_NAME:
    globals()[name] = pd.read_json(data_dir / f"{name}.json", typ='series')
    
# converting all json formats in its raw form to dataframe
# the use of globals is unique here a string variable is assigned as the name of the dataframe
```

```python
# the transformation of the India specific data starts from here
count_indian_artist = len(artists_ids_IN)
df = pd.DataFrame()
for artist_num in range(count_indian_artist):
    df1 = pd.DataFrame(data_artists_IN[artist_num][0]['albums_full'][0])
    df = df.append(df1)
df.reset_index(inplace = True)
df.drop('index', axis=1, inplace = True)
track_data = df['tracks']
df.drop('tracks',axis=1, inplace = True)

df2 = pd.DataFrame()
for row in track_data.array:
    df_indi_track = pd.DataFrame(row)
    df2 = df2.append(df_indi_track)
df2 = df2.reset_index()

features = df2['features']
df2.drop('features',axis=1, inplace = True)

df3 = pd.DataFrame()
for row in features.array:
    df_indi_features = pd.DataFrame(row, index=[0])
    df3 = df3.append(df_indi_features)
df3 = df3.reset_index()

df3.drop('index', axis = 1, inplace = True)
df2.drop('index', axis = 1, inplace = True)
df = df.join(df2)
data_IN = df.join(df3)
data_IN.drop_duplicates(inplace = True)
```
