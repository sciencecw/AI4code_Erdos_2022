```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from textwrap import wrap

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Developer'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['Title']).drop_duplicates()
print(indices)
```

```python
idx = indices['Zwei!!']
print(idx)
```

```python
def get_recommendations(text, cosine_sim=cosine_sim):
    idx = indices[text]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    episodes_indices = [i[0] for i in sim_scores]

    return df['Title'].iloc[episodes_indices]
```

```python
get_recommendations('Zoo Puzzle?SuperLite 2000: ZoooJP')
```
