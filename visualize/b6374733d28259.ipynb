```python

my_list = ['apples', 'berries', 'bananas', 'oranges']

#1  Display all items in my_list via slice notation
#======================================
print(my_list[:])

```

```python

#2  Read a list via Index notation
#======================================
print(my_list[0])
```

```python

#2.1  Get index of item in a list 
#======================================
i = my_list.index('berries')
print(i)
```

```python

#2.2 Display index and items in my_list
#=========================================
for counter, value in enumerate(my_list):
    print(counter, value)
```

```python

#3  Change / Update item in a list
#======================================
my_list [1] = 'cranberries'

#display all items in my_list via slice notation
print(my_list[:])


```

```python

#4  Append item in a list
#======================================
my_list.append('guava') 
#display all items in my_list via slice notation
print(my_list[:])

```

```python

#5  Insert item in a list
#======================================
my_list.insert(len(my_list) +1, 'grapes')
print(my_list[:])

```

```python

#6  Pop item in a list
#======================================
return_value = my_list.pop(3)
print(return_value)
print("My list is now : ", my_list[:])


```

```python

#7  Sort items in a list
#======================================
my_list.sort()
print("Sorted list : " , my_list[:])



```

```python
#8  Find items in a list
#======================================
fruit = 'apples'
if fruit in my_list:
    print("SUCCESS! Found ", fruit, " at index ", my_list.index(fruit))
else:
    print(fruit, " NOT FOUND!")
```

## Nested Lists 

```python

#8  Nested Lists #####################
#======================================

L = ['a', 'b', ['cc', 'dd', ['eee', 'fff']], 'g', 'h']

#List items
for counter, value in enumerate(L):
    print(counter, value)


print("Nested list 'L' : ", str(L))
```

```python
# Pop Quiz : What is the output of the following statement?
# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print(L[2][1])
```

```python
# Pop Quiz : What is the output of the following statement?
# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print(L[2][2][0])
```
