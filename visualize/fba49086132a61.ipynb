##### Imports

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
headers = ['Departments', 'Municipalities', 'DANE_Code', 'Weapons_Medium', 'Date', 'Gender', 'Age_Group', 'No_of_people']
```

```python
df = pd.read_csv('../input/domestic-violence-in-colombia/Reporte_Delito_Violencia_Intrafamiliar_Polic_a_Nacional.csv', low_memory=False, header=None, names=headers)
```

```python
df = df.iloc[1:]
df.columns=headers
```

```python
df.describe()
```

```python
_ = [print(i, df[i].unique(), '\n') for i in df.columns]
```

##### Gender Column

```python
sns.countplot(df.Gender)
```

```python
df.Gender[df.Gender.isnull()]
```

```python
_ = [print(i, '  ', len(df[df.Gender == i])) for i in df.Gender.unique()]
```

```python
df.Gender.replace('-', 'Female', inplace=True)
```

```python
df.Gender.replace('NO REPORTA', 'Female', inplace=True)
```

```python
df.Gender.replace('FEMENINO', 'Female', inplace=True)
```

```python
df.Gender.replace('MASCULINO', 'Male', inplace=True)
```

```python
df.Gender.fillna('Female', inplace=True)
```

```python
_ = [print(i, '  ', len(df[df.Gender == i])) for i in df.Gender.unique()]
```

```python
sns.countplot(x=df.Gender)
plt.show()
```

##### Weapons Medium

```python
df.Weapons_Medium.unique()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df.Weapons_Medium)
plt.xticks(rotation=90)
plt.show()
```

```python
df.Weapons_Medium.replace('ARMA BLANCA / CORTOPUNZANTE', 'Knives', inplace=True)
df.Weapons_Medium.replace('ARMA DE FUEGO', 'Firearm', inplace=True)
df.Weapons_Medium.replace('CONTUNDENTES', 'Hammer/Blunt Weapon', inplace=True)

df.Weapons_Medium.replace('CORTANTES', 'Cutting/Biting/Piercing', inplace=True)
df.Weapons_Medium.replace('CORTOPUNZANTES', 'Sharp Weapon', inplace=True)
df.Weapons_Medium.replace('NO REPORTADO', 'Unreported', inplace=True)
df.Weapons_Medium.replace('PUNZANTES', 'Stabbing', inplace=True)

df.Weapons_Medium.replace('SIN EMPLEO DE ARMAS', 'NO USE OF WEAPONS', inplace=True)
df.Weapons_Medium.replace('NO REPORTA', 'Unreported', inplace=True)
df.Weapons_Medium.replace('ESCOPOLAMINA', 'Drugs/ Colombian Devils Breath', inplace=True)

df.Weapons_Medium.replace('-', 'Hammer/Blunt Weapon', inplace=True)
df.Weapons_Medium.fillna('Hammer/Blunt Weapon', inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=df.Weapons_Medium)
plt.xticks(rotation=15)
plt.show()
```
