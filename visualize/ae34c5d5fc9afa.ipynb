### Required arguments

**Required arguments are the arguments passed to a function in correct positional order. Here, the number of arguments in the function call should match exactly with the function definition.**

```python
def sayhello(helloword):
    
    print(helloword)
    
sayhello("Merhaba")
    
```

### Default arguments

**A default argument is an argument that assumes a default value if a value is not provided in the function call for that argument. The following example gives an idea on default arguments, it prints default age if it is not passed**

```python
def function(x,y = 1):
    
    return x * y

print(function(12)) #we give one parameters but y parameter define default = 2 therefore no errors occurred
```

```python
#what happens we give two parameter
print(function(12,2)) #y value change to 2
```

### Flexible Argument


**You may need to process a function for more arguments than you specified while defining the function. These arguments are called flexible-variable-length  arguments and are not named in the function definition, unlike required and default arguments.**




```python
def function(*args):
    print("İnput parameter",args)
    
    for i in args:
        print("Output parameter",i + 10)
        
    
function(1,2,3,4,5)
```

## List Comprehension


**A list comprehension is a syntactic construct available in some programming languages for creating a list based on existing lists**

**[expression for item in list]**

This is equivalent to:
for item in oldlist:
    newlist.append(expression)

**[expression for item in list if conditional]**


This is equivalent to:
for item in oldlist:
    if conditional:
        newlist.append(expression)


```python
list1 = [1,2,3,4]

list2 = [i+1 for i in list1] #i+1 = expression list1 = oldlist list2 = newlist

print(list2)
```

```python
list1 = [5,11,12,13]

list2 = [i**2 for i in list1 if i > 10] #i**2 = expression if i > 10 condition

print(list2)
```

```python
list1 = [5,6,7,19,18]

list2 = [i**2  for i in list1 if i < 20 if i > 10]
                              #if block #elif
print(list2)
```

### Apply Function

         Diamond Clarity refers to the absence of the Inclusions and Blemishes.
**clarity** a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

**we create new column Color_Level**

**I1,SI2,SI1,VS2 = Bad Clarity**

**VS1,VVS2,VVS1 = Good Clarity**

**IF = Best Clarity**

and use pandas apply method

```python
def seperator(clarity):
    
    if clarity in ["I1","SI2","SI1","VS2"]:
        
        return "Bad Clarity"
    
    elif clarity in ["VS1","VVS2","VVS1"]:
        
        return "Good Clarity"
    
    else:
        
        return "Best Clarity"    
    
data["Clarity_level"] = data["clarity"].apply(seperator)    
    
```

```python
data.head()
```

```python
data["Clarity_level"].value_counts()
```

**Same thing use on 'color' columns**

**D being the best and J the worst.**

**We create new column Color_level**

**D = Best Color**
**E,F,G,H = Good Color**
**H,I,J = Bad Color**

```python
def seperator2(color):
    
    if color in ["E","F","G","H"]:
        return "Good Color"
    elif color in ["H","I","J"]:
        return "Bad Color"
    else:
        return "Best Color"
    

data["Color_level"] = data.color.apply(seperator2)
```

```python
data.head()
```

```python
data["Color_level"].value_counts()
```

```python
condition1  = data.Color_level == "Best Color"
condition2 = data.Clarity_level == "Best Clarity"

bestdiamonds = pd.DataFrame(data[condition1 & condition2])
```
