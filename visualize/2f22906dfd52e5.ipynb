## <u><center> <b> Data Types </u>

### <u> Integer (int) </u>

Integers in python are counting numbers. They do not contain any fractions. They are both positive and negative. 
Eg. -100,-10,-1,0,1,2,10,100 etc..

```python
a = -100
b = 100
print(f'Type of a is {type(a)}')
print(f'Type of b is {type(b)}')
```

```python
age = '22'
print(f'Type of Age is {type(age)}') 
# If using input, Use- age = int(input("Enter Your Age")) to get correct class. This is called explicit type casting.
```

### <u> Floating Numbers (float) </u>

Floating numbers decimal numbers. 
Eg- 1.1,1.2,1001.00,-1.56 etc..

```python
a = 1.1
b = 1001.00
print(f'Type of a is {type(a)}')
print(f'Type of b is {type(b)}')
```

### <u> String (str) </u>

String are characters mostly alphabets. They are written inside single or double quotes. 
Eg- "ABH","XYZ","-1" "1" etc..

```python
a = "GreyWasp"
b = "-1.00"
print(f'Type of a is {type(a)}')
print(f'Type of b is {type(b)}')
```

### <u> Complex Number (complex) </u>

Complex Numbers are numbers that have real and imaginary part.
Eg- 1+3j , 2+0j etc..

```python
a = 1 + 3j
print(f'Type of a is {type(a)}')
b = complex(8)
print(b)
```

## <u><center> Data Structures and their properties </u> </center>

### <u> List </u>
#### Properties:
1. Ordered
2. Mutable
3. Can have multiple data types
4. Can have duplicates
5. Contained in []

```python
list1 = ['1',2,1.198,True,2] #contained in []
print(list1) #Can have multiple data types and duplicates
list1[0] #Ordered
list1[1]=22 #Mutable
print(list1)
print(type(list1))
```

####  Commonly Used Methods 

```python
#append,extend,insert,pop,remove,clear,index,count,sort,sorted,copy,reverse,join,del
list1.append('Name') #Can add only 1 element
print(list1)
list1.append(['xyz','abc']) #If try more, it becomes nested list
print(list1)
list1.pop() #Delete last element
print(list1)
list1.extend(['xyz','abc']) #adding more than 1 element
print(list1)
list1.pop(0) #Delete from specific index
print(list1)
list1.insert(0,45) #Inserting 45 at index 0
print(list1)
list1.remove(45)  #Removing a specific value from list
print(list1)
print(list1.index('xyz')) #To get index of a value
print(list1.count('xyz')) #Count occurence of a value in string
list2  =[25,12,99]
list2.sort() #Sorts list in ascending order and changes list1
print(list2)
print(sorted(list2,reverse=True)) #Sorts list in descending order and but doesn't change list1
print(list2)
list2 = list1.copy() #list2 is same as list1 but changes in list1 won't affect list2. Hard copy
list1.reverse() # Reverses list
print(list1)
list3 = ['a','b','c']
'-'.join(list3) #Join list items by delimeter (only string type of list allowed)
print(list3)
list1.clear() #Deleting elements of list but list is still present
print(list1)
del(list1)#Deletes list1 from system
#print(list1) #Thats why error here
```

### <u> Tuple </u>
####  Properties:
1. Ordered
2. Immutable
3. Can have multiple data types
4. Can have duplicates
5. Contained in ()

```python
tup = (1,2,3,'a',True,1.3,'a') #can have duplicates,multiple data types and enclosed in()
print(type(tup))
print(tup)
#tup[3] = 'Not possible' #Throws error as its immutable
```

####  Commonly Used Methods 

Since, it is immutable, sort or reverse , any function that changes index is not allowed.<br>
Useful for storing latitudes, longitudes etc.

```python
#count,index,len
tup1 = (1,2,3,1,1,1,1)
print(tup.count(1)) #Count no. of occurences of an element
print(tup.index(1)) #Index of first occurence of a element
print(len(tup)) #Length of tuple
```

### <u> Set </u>
#### Properties:
1. Unordered
2. Mutable
3. Can have multiple data types
4. Cannot have duplicates
5. Contained in {}
6. No bool allowed

```python
set1 = {1,1,2,5,6,'a',True} #contained in {}
print(type(set1))
print(set1) #Doesn't inlcude duplicates and has multiple data types . Also, doesn't allow bool
set1.add(100) #Elements are added with add only, mutable though unordered, 100 will be added anywhere
print(set1)
```

####  Commonly Used Methods 

```python
#clear,copy,difference,differece_update,intersection,discard,disjoint,issubset,union,issuperset,remove,symmetric_difference,update,update
#It is just like set in mathematics
A = {i for i in range(5)}
print(A)
B = {i for i in range(1,20,4)}
print(B)
print(A.difference(B)) # No changes to A 
# If we use A.difference_update(B), A gets updated to A-B
print(A.isdisjoint(B))
print(A.issubset(B))
print(A.union(B))
X = A.union(B)
print(X.issuperset(A))
print(A.symmetric_difference(B))#Remove common
A.update([56,78]) #Just like extend
print(A)
```

### <u> Dictionary </u>
#### Properties:
1. Has Key and values (Unordered, has to be accessed by keys)
2. Mutable
3. Can have multiple data types (Keys are mostly strings, but can be tuple or set not list)
4. Overwrites the values for duplicate keys
5. Contained in {}

```python
dict1 = {'Name' : 'GreyWasp',
         'Rank' : 'Kaggle Expert',
         'Age': 21,
        'Rank' : 'Kaggle 1X Expert'} #Multiple data types and overwrite duplicates and contained in {}
print(dict1)
# dict[0] # Wont work as dictionary is not ordered
dict1['Age'] = 22 #Mutable
```

#### Commonly Used Methods

```python
#keys,values,items,pop,popitem,get,fromkeys,copy,update,clear
print(dict1.keys())
print(dict1.values())
print(dict1.items())
dict1.pop("Age") #Delete a key with value
print(dict1)
dict1.popitem() #Deletes last item
print(dict1.get('Age',22)) #Returns 22 if Age not present, else will return value of key 'age'
details = ['Age','Ranking']
print(dict1.fromkeys(details,1))
print(dict1)
```
