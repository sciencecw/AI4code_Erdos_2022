# Data Loading

```python
data=pd.read_csv("../input/machine-learning-24-hrs-hackathon/train_SJC.csv",header=1)
sample=pd.read_csv("../input/machine-learning-24-hrs-hackathon/Test_SJC.csv")
```

```python
data.columns
```

# Data Preprocessing

**Renaming the columns.**

```python
data=data.rename(columns={'Unnamed: 2':'DateReported','Unnamed: 7':'DependentsOther','Unnamed: 11':'DaysWorkedPerWeek'})
```

```python
# Let us drop duplicated rows if any
data.drop_duplicates(inplace=True)
```

**Converting to dates in dataset to datetime objects**

```python
data["DateTimeOfAccident"]=pd.to_datetime(data["DateTimeOfAccident"])
sample["DateTimeOfAccident"]=pd.to_datetime(sample["DateTimeOfAccident"])
```

```python
data["DateReported"]=pd.to_datetime(data["DateReported"])
sample["DateReported"]=pd.to_datetime(sample["DateReported"])
```

**Missing Value treatment**

```python
data["MaritalStatus"]=data["MaritalStatus"].fillna(data["MaritalStatus"].mode()[0])
```

```python
sample["MaritalStatus"]=sample["MaritalStatus"].fillna(sample["MaritalStatus"].mode()[0])
```

```python
sample["MaritalStatus"]=sample["MaritalStatus"].fillna(sample["MaritalStatus"].mode()[0])
```

```python
data["WeeklyWages"]=data["WeeklyWages"].fillna(data["WeeklyWages"].median())
```

```python
data["HoursWorkedPerWeek"]=data["HoursWorkedPerWeek"].fillna(data["HoursWorkedPerWeek"].median())
```

# Feature Extraction

```python
data["YearAcc"]=data["DateTimeOfAccident"].dt.year
data["MonthAcc"]=data["DateTimeOfAccident"].dt.month
data["DayAcc"]=data["DateTimeOfAccident"].dt.day
sample["YearAcc"]=sample["DateTimeOfAccident"].dt.year
```

```python
delay=data["DateReported"]-data["DateTimeOfAccident"]
data["delay_reported"]=[x.days for x in delay]
```

```python
delay1=sample["DateReported"]-sample["DateTimeOfAccident"]
sample["delay_reported"]=[x.days for x in delay1]
```

```python
data['HoursWorkedPerWeek']=np.where(data['HoursWorkedPerWeek']>100,100,data.HoursWorkedPerWeek)
sample['HoursWorkedPerWeek']=np.where(sample['HoursWorkedPerWeek']>100,100,sample.HoursWorkedPerWeek)
```

```python
data["Hourly_wage"]=data['WeeklyWages']/data['HoursWorkedPerWeek']
sample["Hourly_wage"]=sample['WeeklyWages']/sample['HoursWorkedPerWeek']
```

```python
data["Hourly_wage"]=data['WeeklyWages']/data['HoursWorkedPerWeek']
sample["Hourly_wage"]=sample['WeeklyWages']/sample['HoursWorkedPerWeek']
```

```python
data["hoursworkedperday"]=data['HoursWorkedPerWeek']/data["DaysWorkedPerWeek"]
sample["hoursworkedperday"]=sample['HoursWorkedPerWeek']/sample["DaysWorkedPerWeek"]
```

```python
data["Hourly_wage"] = data["Hourly_wage"].replace([np.inf, -np.inf],data["Hourly_wage"].median()).dropna(axis=0)
```

```python
sample["Hourly_wage"] = sample["Hourly_wage"].replace([np.inf, -np.inf],sample["Hourly_wage"].median()).dropna(axis=0)
```

```python
def floor_cap(df,x):
    for i in x:
        low=df[i].quantile(0.05)
        top=df[i].quantile(0.95)
        df[i] = np.where(df[i] <low,low,df[i])
        df[i] = np.where(df[i] >top,top,df[i])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import sklearn.preprocessing as pre
```

```python
df=data[var]
dfs=sample[var1]
```

```python
cat=['Gender','MaritalStatus','PartTimeFullTime']
```

```python
le=LabelEncoder()
for x in cat:
    df[x]=le.fit_transform(df[x])
```

```python
Y=df["UltimateIncurredClaimCost"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
 X_scale=sc.fit_transform(df.drop("UltimateIncurredClaimCost",axis=1))
```

```python
 X_scale1=sc.transform(dfs)
```

```python
import sklearn.model_selection as ms
import sklearn.linear_model as lm
from sklearn.metrics import mean_squared_error
```

```python
x_train,x_test,y_train,y_test=ms.train_test_split(X_scale,Y,test_size=0.3,random_state=1234456)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
glm=lm.LinearRegression()
```

```python
glm.fit(x_train,y_train)
```

```python
glm.score(x_test,y_test)
```

```python
glm.score(x_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_test,glm.predict(x_test)))
```

```python
predictions=glm.predict(X_scale1)
```
