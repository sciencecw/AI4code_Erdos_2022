                             MATPLOTLİB
* Line plot is better when x axis is time.
* Scatter is better when there is correlation between two variables
* Histogram is better when we need to see distribution of numerical data.    

```python
# Line Plot
data.price.plot(kind = 'line', color = 'y',label = 'price',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.points.plot(color = 'g',label = 'points',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('PRICE')             
plt.ylabel('POINTS')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
# Scatter Plot 
data.plot(kind='scatter', x='price', y='points',alpha = 0.5,color = 'red')
plt.xlabel('Price')              # label = name of label
plt.ylabel('Points')
plt.title('Price Points Scatter Plot')  
```

```python
#Histogram
data.price.plot(kind = 'hist',color='red', bins = 50,figsize = (5,5))
plt.show()
```

                            DICTIONARY
* It has 'key' and 'value'
* dictionary = {'Turkiye' : 'Istanbul'}
* Key is Turkiye.
* Values is Istanbul.

```python
dictionary = {'turkiye' : 'istanbul','almanya' : 'berlin'}
print(dictionary.keys())
print(dictionary.values())
```

Lets practice some other properties like keys(), values(), update, add, check, remove key, remove all entries and remove dicrionary.

```python
dictionary['turkiye'] = "sakarya"    # update existing entry
print(dictionary)
dictionary['france'] = "paris"       # Add new entry
print(dictionary)
del dictionary['turkiye']              # remove entry with key 'turkiye'
print(dictionary)
print('france' in dictionary)        # check include or not
dictionary.clear()                   # remove all entries in dict
print(dictionary)
```

WHILE and FOR LOOPS
We will learn the most basic while and for loops

```python
lis = [1,2,3,4,5]
for i in lis:
    print('i : ',i)
print('')
```

```python
for index, value in enumerate(lis): #index : value = 0:1, 1:2, 2:3, 3:4, 4:5
    print(index," : ",value)
print('')  
```

```python
dictionary = {'turkiye':'istanbul','france':'paris'}
for key,value in dictionary.items():
    print(key," : ",value)
print('')
```
