### 0. Sample variable 'json_data'

```python
json_data = {"markers": [
        {
            "point":(40.266044,-74.718479), 
            "homeTeam":"Lawrence Library",
            "awayTeam":"LUGip",
            "markerImage":"images/red.png",
            "information": "Linux users group meets second Wednesday of each month.",
            "fixture":"Wednesday 7pm",
            "capacity":"",
            "previousScore":""
        }]}
```

---
### 1. Save *.json file

```python
import json

with open('./json_file.json','w') as f:
    json.dump(json_data,f)
```

---
### 2. Load *.json file

```python
with open("./json_file.json", "r") as f:
    json_load = json.load(f)

print(json_load)
```

---
### 3. Dump Options

```python
print('[ Original Json Variable ]','-'*20)
print('\n',json_data)

print('[ Indented Json Variable ]','-'*20)
tmp1 = json.dumps(json_data,indent=4)
print('\n',tmp1)

print('[ Indented & Sorted Json Variable ]','-'*20)
tmp2 = json.dumps(json_data,indent=4,sort_keys=True)
print('\n',tmp2)
```
