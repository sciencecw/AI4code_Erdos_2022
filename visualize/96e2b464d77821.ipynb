# Graphing Performance of Stocks

```python
vaccine_stocks['Pfizer_5MA'] = vaccine_stocks.High_BioNTech.rolling(window=5).mean().round(2)
vaccine_stocks.head(5)
Pfizer = vaccine_stocks[['Date','High_BioNTech','Pfizer_5MA']]
```

```python
vaccine_stocks['Astra5MA'] = vaccine_stocks.High_Astrazeneca.rolling(window=5).mean().round(2)
Astra = vaccine_stocks[['Date','High_Astrazeneca','Astra5MA']]
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import figure
```

```python
Pfizer_Max_Date = vaccine_stocks.loc[vaccine_stocks.Pfizer_5MA == vaccine_stocks.Pfizer_5MA.max(), 'Date'] #get date of max price
Pfizer_Min_Date = vaccine_stocks.loc[vaccine_stocks.Pfizer_5MA == vaccine_stocks.Pfizer_5MA.min(), 'Date'] #get date of min price
Pfizer_Max = vaccine_stocks.Pfizer_5MA.max()
Pfizer_Min = vaccine_stocks.Pfizer_5MA.min()

Astra_Min_Date = vaccine_stocks.loc[vaccine_stocks.Astra5MA == vaccine_stocks.Astra5MA.min() ,'Date']
Astra_Max_Date = vaccine_stocks.loc[vaccine_stocks.Astra5MA == vaccine_stocks.Astra5MA.max() ,'Date']
Astra_Min = vaccine_stocks.Astra5MA.min()
Astra_Max = vaccine_stocks.Astra5MA.max()


#convert dates into number
MinNum = mdates.date2num(Pfizer_Min_Date)
MaxNum = mdates.date2num(Pfizer_Max_Date)


MinNumAs = mdates.date2num(Astra_Min_Date)
MaxNumAs = mdates.date2num(Astra_Max_Date)
```

```python
#createsubplot
fig, graphs = plt.subplots(2,1)

#Pfizer
graphs[0].plot(Pfizer.Date,Pfizer.Pfizer_5MA)
graphs[0].set_title("BioNTech Stock Price Over Time", fontsize =16)
graphs[0].set_xlabel("Date", fontsize = 14)
graphs[0].set_ylabel("Price $USD", fontsize = 14)
graphs[0].set_ybound(10,500)
graphs[0].annotate("Low of $13.7",xy=(MinNum,Pfizer_Min),xytext = (MinNum-10,Pfizer_Min+50),fontsize = 18, arrowprops=dict(facecolor='black',shrink = 0.05))
graphs[0].annotate("High of $435.85",xy = (MaxNum,Pfizer_Max),xytext = (MaxNum-50, Pfizer_Max+40),fontsize = 18, arrowprops=dict(facecolor='black',shrink = 0.05))

#AZ
graphs[1].plot(Astra.Date,Astra.Astra5MA)
graphs[1].set_title("AstraZeneca Stock Price Over Time", fontsize = 16)
graphs[1].set_xlabel("Date")
graphs[1].set_ylabel("Price $USD")
graphs[1].set_ybound(38,63)
graphs[1].annotate("Low of $40.64", xy=(MinNumAs, Astra_Min),xytext= (MinNumAs, Astra_Min - 2), fontsize = 18, arrowprops=dict(facecolor='black',shrink = 0.05))
graphs[1].annotate("High of $60.5", xy = (MaxNumAs, Astra_Max), xytext= (MaxNumAs, Astra_Max -8), fontsize = 18, arrowprops=dict(facecolor='black',shrink = 0.05))

#FigConfig
plt.subplots_adjust(top=1.5)
fig.set_size_inches(20,10)

fig.show()
```
