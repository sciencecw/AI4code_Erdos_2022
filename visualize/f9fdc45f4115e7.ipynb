```python
df = pd.read_csv('/kaggle/input/person-attributes/dataset_person_attributes_train.csv')
```

```python
# height(feet),weight(lbs), foot size(inches)
df.head(10)
```

```python
df.columns
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
train_cols = ['height', 'weight', 'foot size']
response_col = ['Person']
```

```python
clf = GaussianNB()
clf.fit(df[train_cols], df[response_col])
```

```python
df_test = pd.read_csv('/kaggle/input/person-attributes/dataset_person_attributes_test.csv')
df_test.head()
```

```python
clf.predict(df_test[train_cols])
```

```python
np.round(clf.predict_proba(df_test[train_cols]),2)
```

<font style="font-size:15px;color:blue;font-style: italic;" >
