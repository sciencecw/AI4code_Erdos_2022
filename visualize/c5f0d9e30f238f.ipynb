Mendel's Pea Flowers

Mendel's model can be formulated as a hypothesis that we can test.

1. Null Hypothesis. For every plant, there is a 75% chance that it will have purple flowers, and
a 25% chance that the flowers will be white, regardless of the colors in all the other plants.

2. Alternative Hypothesis. Mendel's model isn't valid.

## Create data set

```python
import random

flowers = ['purple']*705
white = ['white']*224

flowers.extend(white)

random.shuffle(flowers)
flowers[700:710]
```

## Convert list to CSV File

```python
import pandas as pd

# Read csv file 
df = pd.read_csv("../input/mendels-pea-flower-distribution/Mendel_dataset.csv")
df.describe()
```

```python
# Calculate no of flowers which are purple
purple_count = df[df["Flower"]  == 'purple'].count()
print("Purple flowers count is :")
purple_count
```

```python
# Calculate no of flowers which are white
white_count = df[df["Flower"]  == 'white'].count()
print("White flowers count is :")
white_count
```

```python
# Calculate total no of flowers in dataset
total_flowers = df["Flower"].count()
print("Total Flower count :", total_flowers)
```

```python
observed_proportion = purple_count/total_flowers
observed_proportion
```

Test Statistic. Since the model predicts 75% purple-flowering plants, a relevant statistic is
the difference between 0.75 and the observed proportion of purple-flowering plants.

```python
observed_statistic = abs(observed_proportion - 0.75)
observed_statistic
```

we need use a model to simulate new samples of plants and calculate the statistic for each one.

```python
import numpy as np

# create a new model to simulate new samples.
model_colors = ['purple', 'purple', 'purple', 'white']
new_sample = np.random.choice(model_colors, total_flowers)
```

To compare with our observed statistic, we need to know by how much the proportion of
purple-flowering plants in this new sample differs from 0.75.

```python
proportion_purple = np.count_nonzero(new_sample == 'purple')/total_flowers
print("Proportion of purple-flowers in this new sample which differs from 0.75 is", abs(proportion_purple - 0.75))
```

## Empirical distribution of the test statistic, if the null hypothesis were true.

```python
import numpy.random as npr

repetitions = 5000

for i in range(repetitions):
    sim = npr.binomial(total_flowers, 0.75, repetitions)
    
sim
```

```python
df = pd.DataFrame(sim, columns =['Count'])

df['sampled_stats']= (df['Count'] / total_flowers)
df = df.set_index('Count')
df['Distance from 0.75'] = abs(df['sampled_stats'] - 0.75)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['Distance from 0.75'])
plt.axvline(x = 0.008881, color = 'magenta')
plt.title('Empirical distribution')
plt.xlabel('Distance from 0.75')
plt.ylabel('Percent Per Unit')
plt.grid()
plt.show()
```

Conclusion of the test. The observed value of the statistic, based on Mendel's data, is
0.008881 which is just under 0.01. That's right in the heart of this distribution.

The statistic based on Mendel's data is consistent the distribution that our simulation
predicts based on Mendel's model. Thus the data support the null hypothesis – Mendel's
model is good – more than they support the alternative.
