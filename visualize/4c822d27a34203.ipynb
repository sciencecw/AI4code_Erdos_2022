# Predicting Amsterdam House Prices with MLR

#### Loading and Displaying data

```python
import pandas as pd
```

```python
df = pd.read_csv("../input/amsterdam-house-price-prediction/HousingPrices-Amsterdam-August-2021.csv")

df.head()
```

#### Dropping irrelevant column and transforming column values

```python
df["Zip"] = df["Zip"].str.extract('(\d+)', expand=False)
df = df.drop(["Address"], axis  = 1)
df = df.dropna()

df.head()
```

#### Choosing parameters

```python
X = df.drop(labels = ["Price"], axis = 1).values

y = df[["Price"]].values
```

#### Splitting data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

#### Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
