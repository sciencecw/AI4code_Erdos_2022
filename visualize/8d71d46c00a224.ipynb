```python
# packages
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# load data / preview
df = pd.read_csv('../input/google-trends-dataset/trends.csv')
df.head(10)
```

# General EDA

```python
# eval distributions
features_general = ['year','location', 'query', 'category','rank']

for f in features_general:
    print(df[f].value_counts())
    # plot top 25
    plt.figure(figsize=(10,4))
    df[f].value_counts()[0:25].plot(kind='bar')
    plt.title(f)
    plt.grid()
    plt.show()
```

# Filter specific region

```python
# filter specific region
sel_region = 'United States'
df_filter = df[df.location==sel_region]
```

```python
features = ['year','category','query']

for f in features:
    plt.figure(figsize=(10,4))
    df_filter[f].value_counts()[0:25].plot(kind='bar')
    plt.title('Location = ' + sel_region + ' / ' + f)
    plt.grid()
    plt.show()
```

<a id='3'></a>
# Filter specific year

```python
# filter specific year
sel_year = 2020
df_filter = df[df.year==sel_year]
```

```python
features = ['location', 'category', 'query']

for f in features:
    plt.figure(figsize=(10,4))
    df_filter[f].value_counts()[0:25].plot(kind='bar')
    plt.title('Year = ' + str(sel_year) + ' / ' + f)
    plt.grid()
    plt.show()
```
