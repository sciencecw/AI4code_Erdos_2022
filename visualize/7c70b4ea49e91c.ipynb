## Most common last names in dataset

First we should seperate first name and last name.
we can do this in two ways.

seprating first name and last name by space. but this cause a problem for name strings with more than one space. so we have to set a condition to exlude them.

```python
df[['first_name','last_name']] = df['name'].loc[df['name'].str.split().str.len() == 2].str.split(expand=True)
df['last_name'].value_counts().head()
```

another way for finding last names is to split names by space and assume last elemnt is last name.

```python
df['name'].str.split().str[-1].value_counts().head()
```
