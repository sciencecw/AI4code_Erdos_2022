```python
from colorama import Fore, Style

nRowsRead = 1000 # specify 'None' if want to read whole file
# ham_lyrics.csv has 3634 rows in reality, but we are only loading/previewing the first 1000 rows
df = pd.read_csv('../input/lacountyrestaurantinspectionandviolations/violations.csv', delimiter=',', nrows = nRowsRead)
df.dataframeName = 'violations.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
print(Fore.YELLOW + 'Data shape: ',Style.RESET_ALL,df.shape)
df.head()
```

#Spoiler alert: No Missing Values.

```python
#Code from Gabriel Preda
#plt.style.use('dark_background')
def plot_count(feature, title, df, size=1):
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette=('#42b3f5', '#42f554', '#f55442'))
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

```python
plot_count("points", "Violation Points", df,2)
```

```python
plot_count("serial_number", "Serial Number", df,4)
```

```python
plot_count("violation_code", "Violation Code", df,4)
```

```python
plot_count("violation_status", "Violation Status", df,1)
```

```python
plot_count("violation_description", "violation_description", df,4)
```

```python
#Code by Taha07  https://www.kaggle.com/taha07/data-scientists-jobs-analysis-visualization/notebook

color = plt.cm.Greens(np.linspace(0,1,20))
df["violation_description"].value_counts().sort_values(ascending=False).head(20).plot.pie(y="points",colors=color,autopct="%0.1f%%")
plt.title("Restaurant Violations Descriptions")
plt.axis("off")
plt.show()
```

```python
# Count Plot
plt.style.use("classic")
plt.figure(figsize=(10, 8))
sns.countplot(df['violation_description'], palette='Greens', **{'hatch':'/','linewidth':3})
plt.xlabel("violation_description")
plt.ylabel("Count")
plt.title("Restaurants Violations")
plt.xticks(rotation=45, fontsize=8)
plt.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.violation_description)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Oranges', background_color="green").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
