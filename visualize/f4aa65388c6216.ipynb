# Create Linear Regression to  predict Sales Prices

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
from sklearn.linear_model import LinearRegression


#Read Data
traindata = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
testdata=pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')

#Clean Data "nan"
testdata['LotArea'] = testdata['LotArea'].fillna(0)
testdata['TotalBsmtSF'] = testdata['TotalBsmtSF'].fillna(0)
testdata['GrLivArea'] = testdata['GrLivArea'].fillna(0)

#specify feature
data=['LotArea','TotalBsmtSF','GrLivArea']
X_train = traindata.iloc[:, traindata.columns.isin(list(data))]
y_train= traindata.iloc[:,-1]
X_test = testdata. iloc[:, testdata.columns.isin(list(data))]
y_test = testdata. iloc[:, -1]

#Start Linear Regression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
regressor.score(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
y_pred_arr=np.array(y_pred)[np.newaxis]
id_arr=np.array(testdata['Id'])[np.newaxis]
c=np.column_stack((id_arr.T,y_pred_arr.T))
df = pd.DataFrame(c)
print(df)
```
