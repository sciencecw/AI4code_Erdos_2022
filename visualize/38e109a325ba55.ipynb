```python
data = pd.read_csv('../input/iris/Iris.csv')
```

```python
X = data.drop(['Species','Id'],axis = 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

> #### KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
km_3 = KMeans(n_clusters = 3).fit(X)
X['Cluster'] = km_3.labels_
```

```python
X.head()
```

```python
data.head()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = X['PetalLengthCm'],y = X['PetalWidthCm'],hue = X['Cluster'],palette = 'bright')
plt.title('KMeans with 3 clusters')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = data['PetalLengthCm'],y = data['PetalWidthCm'],hue = data['Species'],palette = 'bright')
plt.title('Different Species')
plt.show()
```

> #### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
AC = AgglomerativeClustering(n_clusters = 3,linkage = 'average').fit(X)
X['Cluster_AC'] = AC.labels_
```

```python
X.head()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = X['PetalLengthCm'],y = X['PetalWidthCm'],hue = X['Cluster_AC'],palette = 'bright')
plt.title('Agglomerative Clustering with 3 clusters')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = data['PetalLengthCm'],y = data['PetalWidthCm'],hue = data['Species'],palette = 'bright')
plt.title('Different Species')
plt.show()
```

> #### DBScan

```python
from sklearn.cluster import DBSCAN 
```

```python
db= DBSCAN(eps=1, min_samples=6).fit(X)
X['Cluster_DB'] = db.labels_
```

```python
X.head()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = X['PetalLengthCm'],y = X['PetalWidthCm'],hue = X['Cluster_DB'],palette = 'bright')
plt.title('DBScan')
plt.show()
```
