# import libraries

```python
import numpy as np
import pandas as pd 
```

# load data 

```python
train = pd.read_csv("../input/titanic/train.csv",index_col = "PassengerId")
```

# make dataframe with first 5 data

```python
start5 = train[:5]
start5
```

# make dataframe with last 5 data

```python
end5 = train[-5:]
end5
```

# concat dataframes 

Concatenate pandas objects along a particular axis with optional set logic along the other axes.

Can also add a layer of hierarchical indexing on the concatenation axis, which may be useful if the labels are the same (or overlapping) on the passed axis number.

## maintain index 

```python
train_sub = pd.concat([ start5,end5 ])
train_sub
```

## delete index

```python
train_sub = pd.concat([ start5,end5 ],ignore_index =True)
train_sub
```
