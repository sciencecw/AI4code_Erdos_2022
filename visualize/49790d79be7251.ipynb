# Importing and Merging Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Converting some binary variables (Yes/No) to 0/1
varlist =  ['PhoneService', 'PaperlessBilling', 'Churn', 'Partner', 'Dependents']

# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the housing list
telecom[varlist] = telecom[varlist].apply(binary_map)
```

```python
# creating dummy varaibles
dummy1 = pd.get_dummies(telecom[['Contract', 'PaymentMethod', 'gender', 'InternetService']], drop_first=True)

# Adding the results to the master dataframe
telecom = pd.concat([telecom, dummy1], axis=1)
```

```python
ml = pd.get_dummies(telecom['MultipleLines'], prefix='MultipleLines')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['MultipleLines_No phone service'], 1)
#Adding the results to the master dataframe
telecom = pd.concat([telecom,ml1], axis=1)


os = pd.get_dummies(telecom['OnlineSecurity'], prefix='OnlineSecurity')
os1 = os.drop(['OnlineSecurity_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,os1], axis=1)

# Creating dummy variables for the variable 'OnlineBackup'.
ob = pd.get_dummies(telecom['OnlineBackup'], prefix='OnlineBackup')
ob1 = ob.drop(['OnlineBackup_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,ob1], axis=1)

# Creating dummy variables for the variable 'DeviceProtection'. 
dp = pd.get_dummies(telecom['DeviceProtection'], prefix='DeviceProtection')
dp1 = dp.drop(['DeviceProtection_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,dp1], axis=1)

ts = pd.get_dummies(telecom['TechSupport'], prefix='TechSupport')
ts1 = ts.drop(['TechSupport_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,ts1], axis=1)

# Creating dummy variables for the variable 'StreamingTV'.
st =pd.get_dummies(telecom['StreamingTV'], prefix='StreamingTV')
st1 = st.drop(['StreamingTV_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,st1], axis=1)

# Creating dummy variables for the variable 'StreamingMovies'. 
sm = pd.get_dummies(telecom['StreamingMovies'], prefix='StreamingMovies')
sm1 = sm.drop(['StreamingMovies_No internet service'], 1)
# Adding the results to the master dataframe
telecom = pd.concat([telecom,sm1], axis=1)
```

```python
# We have created dummies for the below variables, so we can drop them
telecom = telecom.drop(['Contract','PaymentMethod','gender','MultipleLines','InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies'], 1)
```

# Trying with Differnet Models to get better accuracy

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif.head()
```

```python
features_to_remove = vif.loc[vif['VIF'] >= 4.99,'Features'].values
features_to_remove = list(features_to_remove)
print(features_to_remove)
```

```python
X_train = X_train.drop(columns=features_to_remove, axis = 1)
X_train.head()
```

```python
X_test = X_test.drop(columns=features_to_remove, axis = 1)
X_test.head()
```

# Evaluating Model

```python
# Evaluting model using precision_score, recall_score, accuracy_score,f1_score
from sklearn.metrics import precision_score, recall_score, accuracy_score,f1_score
```

```python
y_test_pred_svm=svm_model.predict(X_test)
```

```python
y_test_pred_naivebayes=naivebayes_model.predict(X_test)
```

```python
y_test_pred_dt=dt_model.predict(X_test)
```

```python
y_test_pred_rft=rft_model.predict(X_test)
```

```python
y_test_pred_gbm=gbm_model.predict(X_test)
```

```python
y_test_pred_xgb=xgb_model.predict(X_test)
```

```python
print("SVM precision Score", precision_score(y_test,y_test_pred_svm))
print("naivebayes precision Score", precision_score(y_test,y_test_pred_naivebayes))
print("Decision Tree precision Score", precision_score(y_test,y_test_pred_dt))
print("Random Forest precision Score", precision_score(y_test,y_test_pred_rft))
print("GBM precision Score", precision_score(y_test,y_test_pred_gbm))
print("XGB precision Score", precision_score(y_test,y_test_pred_xgb))
```

```python
print("SVM recall Score", recall_score(y_test,y_test_pred_svm))
print("naivebayes recall Score", recall_score(y_test,y_test_pred_naivebayes))
print("Decision Tree recall Score", recall_score(y_test,y_test_pred_dt))
print("Random Forest recall Score", recall_score(y_test,y_test_pred_rft))
print("GBM recall Score", recall_score(y_test,y_test_pred_gbm))
print("XGB recall Score", recall_score(y_test,y_test_pred_xgb))
```

```python
print("SVM accuracy Score", accuracy_score(y_test,y_test_pred_svm))
print("naivebayes accuracy Score", accuracy_score(y_test,y_test_pred_naivebayes))
print("Decision Tree accuracy Score", accuracy_score(y_test,y_test_pred_dt))
print("Random Forest accuracy Score", accuracy_score(y_test,y_test_pred_rft))
print("GBM accuracy Score", accuracy_score(y_test,y_test_pred_gbm))
print("XGB accuracy Score", accuracy_score(y_test,y_test_pred_xgb))
```

```python
print("SVM F1 Score", f1_score(y_test,y_test_pred_svm))
print("naivebayes F1 Score", f1_score(y_test,y_test_pred_naivebayes))
print("Decision Tree F1 Score", f1_score(y_test,y_test_pred_dt))
print("Random Forest F1 Score", f1_score(y_test,y_test_pred_rft))
print("GBM F1 Score", f1_score(y_test,y_test_pred_gbm))
print("XGB F1 Score", f1_score(y_test,y_test_pred_xgb))
```
