# Create new columns based on age called age_group

age<12 call child

age<20 call it teen

age>=20 and age <=50 call it adult

age>50 call it senior

```python
def categorise_age(age):
    if age <= 12 : 
        return 'child'
    if age <= 20 : 
        return 'teen'
    if age >50 :
        return 'senior'
    else :
        return 'adult'

train['age_group']=train.Age.apply(lambda x :  categorise_age(x))
```

```python
train.age_group.value_counts()
```
