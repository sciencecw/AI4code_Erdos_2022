```python
def preprocess_inputs(df):
    df = df.copy()
    
    # Binary encoding
    df['Gender'] = df['Gender'].replace({
        'Female': 0,
        'Male': 1
    })
    df['Customer Type'] = df['Customer Type'].replace({
        'disloyal Customer': 0,
        'Loyal Customer': 1
    })
    df['Type of Travel'] = df['Type of Travel'].replace({
        'Personal Travel': 0,
        'Business travel': 1
    })
    df['Class'] = df['Class'].replace({
        'Eco': 0,
        'Eco Plus': 1,
        'Business': 2
    })
    
    return df
```
