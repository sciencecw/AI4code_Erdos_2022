Contoh implementasi stack dalam kehidupan sehari hari adalah sebagai contoh saat kita Mengambil buku di perpustakaan yang tertumpuk. Maka kita akan mengambil buku dari yang paling belakang sampai buku pertama atau awal.

```python
tumpukan = [1,2,3,4,5,6,7,8]
print('data sekarang: ',tumpukan)

# memasukkan data baru
tumpukan.append(9)
print('data masuk; ',tumpukan)
tumpukan.append(10)
print('data masuk: ',tumpukan)

out = tumpukan.pop()
print('data keluar: ',out)
print('data sekarang: ',tumpukan)
```

Contoh implementasi queue dalam kehidupan sehari hari adalah sebagai contoh saat kita pergi membeli Makanan di McDonald'S dan ada antrian beberapa orang dihitung dari orang pertama hingga paling belakang atau terakhir.

```python
from collections import deque

antrian = deque( [1,2,3,4,5])

print('orang sekarang: ',antrian)

#menambahkan orang
antrian.append(6)
print('orang masuk: ',6)
print('orang sekarang: ',antrian)

antrian.append(7)
print('orang masuk: ',7)
print('orang sekarang: ',antrian)

antrian.append(8)
print('orang masuk: ',8)
print('orang sekarang: ',antrian)

#mengurangi antrian orang
out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)

out = antrian.popleft()
print('orang keluar: ',out)
print('orang sekarang: ',antrian)
```
