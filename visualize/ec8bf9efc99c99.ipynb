```python
name=[]
region=[]
subregion=[]
population=[]
demonym=[]
area=[]
gini=[]
population=[]
num_code=[]
capital=[]
for i in range(len(data)):
    name.append(data[i]['name'])
    region.append(data[i]['region'])
    subregion.append(data[i]['subregion'])
    population.append(data[i]['population'])
    demonym.append(data[i]['demonym'])
    area.append(data[i]['area'])
    try:
        gini.append(data[i]['gini'])
    except:
        gini.append('None')
    num_code.append(data[i]['numericCode'])
    capital.append(data[i]['capital'])
```

```python
df=pd.DataFrame({
    'name':name,
    'region':region,
    'subregion':subregion,
    'population':population,
    'demonym':demonym,
    'area':area,
    'gini':gini,
    'num_code':num_code,
    'capital':capital,
})
```

### 8.Real Growth Rating

```python
RGR=[]
for i in range(len(name)):
    x=name[i]
    lst=[]
    for j in range(len(x)):
        lst.append(x[j])
        if x[j]==' ':
            lst.pop()
            lst.append('-')
    x=''.join(lst)
    k=0
    c=0
    webpage=requests.get('https://www.infoplease.com/world/countries/{}'.format(x)).text
    soup=BeautifulSoup(webpage,'lxml')
    soup.prettify()
    print(name[i])
    if name[i]=='Gambia' or name[i]=='Palau' or name[i]=='South Africa' or name[i]=='Western Sahara':
        c=20
        
    try:
        while c!=20:
            
            try:
                if soup.find_all('li')[118].find_all('p')[k].find_all('b')[2].text.strip()=='Real growth rate:':
                    x=soup.find_all('li')[118].find_all('p')[k].find_all('b')[2].next_sibling.strip()
                    RGR.append(x)
                    break
            except:
                k=k+1
                c=c+1
    except:
        pass
    if c==20:
        RGR.append(np.nan)
```

```python
RGR
```

```python
len(RGR)
```

```python
df['Real Growth Rating(%)']=RGR
```
