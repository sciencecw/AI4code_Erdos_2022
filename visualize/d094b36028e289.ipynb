# get folder name => Target 

```python
def getLastFolderName(x): 
    pathwithoutFilename = os.path.split(x)[0] 
    lastFolderName = pathwithoutFilename.split("/")[-1] 
    return lastFolderName

targets = list(map(lambda x: getLastFolderName(x),filepaths)) 
targets[:5]
```

# make train dataframe

```python
column_Imagepath = pd.Series(filepaths, name='ImagePath').astype(str)
column_ImageType = pd.Series(targets, name='ImageType').astype(str)

train = pd.merge(column_Imagepath, column_ImageType, right_index = True, left_index = True)
train= train.sample(frac = 1).reset_index(drop=True) # all shuffle 
train.head()
```

# draw images

```python
f,a = plt.subplots(nrows=4, ncols=4,figsize=(15, 10),
                        subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(a.flat):
    ax.imshow(plt.imread(train.ImagePath[i]))
    ax.set_title(train.ImageType[i])
    
plt.tight_layout()
plt.show()
```
