#### 4- Show Domain of Mail Address (Function)

```python
def maildomain(mailaddress):
    if "@" not in mailaddress: return('Error: Wrong email address')
    x = mailaddress.split('@')
    return x[1]
mail = input('Please Enter Your Email Address:')    
domain = maildomain(mail)
print('Domain of your Email Address is: {}'.format(domain))
```

#### 6- Traffic Offense Function

```python
def Ticket(speed):
    birthday = input('Is today your birthday?(yes/no)')
    if birthday == 'yes':
        gift = 5
    else:
        gift = 0
    if speed <= (60+gift):
        return ('No Ticket')
    if speed >= 60 and speed <=(80+gift) :
        return ('Small Ticket')
    if speed >= (80+gift):
        return ('Big Ticket')
Ticket(64)
```
