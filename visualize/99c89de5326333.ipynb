```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from skompiler import skompile
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('../input/fish-market/Fish.csv')
df.rename(columns= {'Length1':'LengthVer', 'Length2':'LengthDia', 'Length3':'LengthCro'}, inplace=True)
df
```

```python
X = df.drop(['Species'], axis=1)
y = df['Species']
```

```python
y = pd.get_dummies(y, columns = ['Species'], drop_first=False)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state =42)
```

```python
cart_model = DecisionTreeClassifier(random_state=42).fit(X_train,y_train)
```

```python
cart_model.max_features_ , cart_model.feature_importances_
```

```python
cart_model.get_depth() , cart_model.get_n_leaves() , cart_model.get_params()
```

```python
cart_model.score(X_train , y_train) , cart_model.score(X_test, y_test)
```

### optimum model için kural yapısını görelim

```python
print(skompile(cart_model.predict).to("python/code"))
```

### Model Tuning işlemleri

```python
cart_grid = { "max_depth" : range(1,11),
            "min_samples_split": range(2,50)}
```

```python
cart1 = DecisionTreeClassifier()
```

```python
cart_cv = GridSearchCV(cart1, cart_grid, cv=10, n_jobs=-1, verbose = 2)
```

```python
cart_cv_model = cart_cv.fit(X_train, y_train)
```

```python
cart_cv_model.best_params_
```

```python
cart1 = DecisionTreeClassifier(max_depth= 7, min_samples_split= 4).fit(X_train,y_train)
```

```python
cart1.score(X_train, y_train) , cart1.score(X_test, y_test)
```
