# PDF paper is friendly to human but it is not easy to read by computer.
In this code, I introduce how to extract sections from PDF by using [Science Parse](https://github.com/allenai/science-parse)

## Download jar file from science parse repository

```python
!wget https://github.com/allenai/science-parse/releases/download/v2.0.3/science-parse-cli-assembly-2.0.3.jar
```

## Specifying input

This tool takes input as positional parameters, in this case, specify path to pdf folder. After type command, extracting start (it takes long time).

```python
!mkdir json
!java -Xmx6g -jar science-parse-cli-assembly-2.0.3.jar ../input/cvpr-2019-papers/CVPR2019/papers -o ./json
```

After finishing, sections files (json format) store in json folder you make. 

```python
from glob import glob
jsons = glob("./json/*")
print("Num of output jsons:", len(jsons))
```
