Last month(September) TPS dataset was ~1GB and this month(October) TPS dataset is ~3GB.

So, this makes clear sense that we need a mechanism to load dataset faster.

There are multiple frameworks which offer faster load time than pandas. In this notebook, I am sharing a simple usage of datatable.

**October TPS Dataset Load Time**
1. Datatable - ~25 secs
2. Pandas - ~90 secs

One advantage I see with datatable is that we can convert the datatable dataframe into pandas dataframe and continue to have fun with pandas as we all are familiar with. This decreases the learning curve to adapt a completely new library to handle data.

# Imports

```python
import numpy as np 
import pandas as pd
import datatable as dt
```

# Load dataset using datatable

```python
%%time
train_dt = dt.fread('../input/tabular-playground-series-oct-2021/train.csv').to_pandas()
test_dt  = dt.fread('../input/tabular-playground-series-oct-2021/test.csv').to_pandas()
sub_dt   = dt.fread('../input/tabular-playground-series-oct-2021/sample_submission.csv').to_pandas()
```

# Load dataset using Pandas

```python
%%time
train_pd = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
test_pd = pd.read_csv('../input/tabular-playground-series-oct-2021/test.csv')
sub_pd = pd.read_csv('../input/tabular-playground-series-oct-2021/sample_submission.csv')
```
