```python
df.drop(['total_votes','rank','start','trend'], axis=1,inplace=True)
```

```python
df.info()
```

```python
#Time skip in one piece happens near 516-517 episode number
#lets check the average rating for all 958 episodes
k=0
for i in df['average_rating']:
    k=k+i
g=len(df['episode'])
print('Average rating for all episode:',k/g)
```

```python
#lets check the average rating for the first 516 episodes
df1=df.iloc[0:516]
df1
```

To find average rating of first 516 episodes

```python
l=0
for j in df1['average_rating']:
    l=l+j
h=len(df1['episode'])
print('Average rating for the first 516 episodes:',l/h)
```

```python
#lets check the average rating after the time skip from 516 episodes
df2=df.iloc[516:958]
df2
```

To find average rating after the time skip

```python
m=0
for z in df2['average_rating']:
    m=m+z
x=len(df2['episode'])
print('Average rating after time skip:',m/x)
```

Seems, even after the time skip the average rating for anime has not decreased but rather increased.
