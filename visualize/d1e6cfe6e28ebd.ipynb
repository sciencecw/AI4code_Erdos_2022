**BASIC EDA**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(df['Education'],palette='rocket_r')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(df['JoiningYear'],palette='rocket_r')
```

* a lot of people joined in 2017

```python
plt.figure(figsize=(6, 6))
classx = ['Delhi','Bangalore','Pune']
colors = sns.color_palette("hls", 8)[5:8]
countx = [len(df[df.City == 'Bangalore']),len(df[df.City == 'New Delhi']),len(df[df.City == 'Pune'])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

* most of the employee are working in delhi

```python
plt.figure(figsize=(6, 6))
classx = ['1','2','3']
colors = sns.color_palette("hls", 8)[5:8]
plt.title('Payment tier')
countx = [len(df[df.PaymentTier == 1]),len(df[df.PaymentTier == 2]),len(df[df.PaymentTier == 3])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.histplot(df['Age'])
plt.show()
```

```python
plt.figure(figsize=(6, 6))
classx = ['Male','Female']
colors = sns.color_palette("hls", 8)[6:8]
plt.title('Gender')
countx = [len(df[df.Gender == 'Male']),len(df[df.Gender == 'Female'])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
classx = ['Yes','No']
colors = sns.color_palette("hls", 8)[6:8]
plt.title('Benched')
countx = [len(df[df.EverBenched == 'Yes']),len(df[df.EverBenched == 'No'])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(df['ExperienceInCurrentDomain'],palette='rocket_r')
plt.show()
```

* very less employee have experience over 5 years

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(df['LeaveOrNot'],palette='magma')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(data=df,x='LeaveOrNot',hue='Education')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(data = df,x='City',hue='LeaveOrNot')
plt.show()
```

* Most people in bangalore don't leave their job

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(data=df,x='PaymentTier',hue='LeaveOrNot')
plt.show()
```

* people with payment tier 2 have higher chance of leaving

```python
fig,ax = plt.subplots(figsize=(10,7))
sns.countplot(data=df,x='Age',hue='LeaveOrNot',palette='rocket')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
classx = ['Male','Female']
colors = sns.color_palette("hls", 8)[6:8]
dfx = df[df.LeaveOrNot == 1]
countx = [len(dfx[dfx.Gender == 'Male']),len(dfx[dfx.Gender == 'Female'])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

* Females have higher chance of leaving

```python
plt.figure(figsize=(6, 6))
classx = ['Benched','Not benched']
colors = sns.color_palette("hls", 8)[6:8]
dfx = df[df.LeaveOrNot == 1]
countx = [len(dfx[dfx.EverBenched == 'Yes']),len(dfx[dfx.EverBenched == 'No'])]
plt.pie(countx, labels = classx,colors=colors,autopct='%1.2f%%')
plt.show()
```

* Very less people who leave the company were benched

```python
dfx = df[df.Gender == 'Female']
fig,ax = plt.subplots(figsize=(10,7))
plt.title('FEMALES IN DIFFERENT CITIES')
sns.countplot(data = dfx,x='City',hue='LeaveOrNot',palette='flare')
plt.show()
```

* A lot of females working in pune leave the company
