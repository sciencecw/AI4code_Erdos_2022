```python
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(olp['Gold'])
plt.title('Distribution of Gold Medals')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(olp['Silver'])
plt.title('Distribution of Silver Medals')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(olp['Bronze'])
plt.title('Distribution of Bronze Medals')
plt.show()
```

```python
import plotly.express as px
fig = px.line(olp['Total'])
fig.show()
```

```python
sns.pairplot(olp)
plt.show()
```

```python
g = sns.PairGrid(olp, hue="Total")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()

plt.show()
```

```python
g = sns.PairGrid(olp)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False)
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(olp['Gold'],labels=olp['Team'])
plt.show()
```

```python
med1 = go.Bar(
    x=olp['Team'],  # Team stands for National Olympic Committee
    y=olp['Gold'],
    name = 'Gold',
    marker=dict(color='#FFD700') # set the marker color to gold
)
med2 = go.Bar(
    x=olp['Team'],
    y=olp['Silver'],
    name='Silver',
    marker=dict(color='#9EA0A1') # set the marker color to silver
)
med3 = go.Bar(
    x=olp['Team'],
    y=olp['Bronze'],
    name='Bronze',
    marker=dict(color='#CD7F32') # set the marker color to bronze
)
data = [med1, med2, med3]
layout = go.Layout(
    title='2018 Winter Olympic Medals by Country'
)
fig = go.Figure(data=data, layout=layout)
fig.show()
pyo.plot(fig, filename='Tokyo.html')
```

```python
med1 = go.Bar(
    x=olp['Team'], 
    y=olp['Gold'],
    name = 'Gold',
    marker=dict(color='#FFD700')
)
med2 = go.Bar(
    x=olp['Team'],
    y=olp['Silver'],
    name='Silver',
    marker=dict(color='#9EA0A1') 
)
med3 = go.Bar(
    x=olp['Team'],
    y=olp['Bronze'],
    name='Bronze',
    marker=dict(color='#CD7F32') 
)
data = [med3, med2, med1]
layout = go.Layout(
    title='2020 Tokyo Olympic Medals by Country',
    barmode='stack'
)
fig = go.Figure(data=data, layout=layout)
fig.show()
pyo.plot(fig, filename='olympic2020.html')

```
