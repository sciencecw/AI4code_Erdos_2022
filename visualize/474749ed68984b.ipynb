## This is the basic Data Visualization of American Top College

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df=pd.read_csv('../input/forbes-americas-top-colleges-2019/ForbesAmericasTopColleges2019.csv')
df.head()
```

```python
print(df.info())
df.describe()
```

## Data Cleaning

```python
df['Net Price'].fillna(value=df['Net Price'].mean(),inplace=True)
df['Average Grant Aid'].fillna(value=df['Average Grant Aid'].mean(),inplace=True)
df['Alumni Salary'].fillna(df['Alumni Salary'].mean(),inplace=True)
df['Acceptance Rate'].fillna(df['Acceptance Rate'].mean(),inplace=True)
df['SAT Lower'].fillna(df['SAT Lower'].mean(),inplace=True)
df['SAT Upper'].fillna(df['SAT Upper'].mean(),inplace=True)
df['ACT Lower'].fillna(df['ACT Lower'].mean(),inplace=True)
df['ACT Upper'].fillna(df['ACT Upper'].mean(),inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

## Data Visualization

```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.columns
```

### Number of college in different state difference with Public and Private College

```python
fig = px.bar(df, x="State", barmode="group",color='Public/Private')
fig.show()
```

### Number of Public and Private college in America

```python
fig=px.bar(df,x='Public/Private')
fig.show()
```

### Scatter plot of using different features 

```python
fig=px.scatter_matrix(df, dimensions=["Net Price", "Average Grant Aid", "Total Annual Cost", "Alumni Salary"], color="Public/Private")
fig.show()
```

### Scatter plot of Alumni Salary with difference of Public and Private College

```python
fig=px.scatter(df, x="Alumni Salary", color="Public/Private")
fig.show()
```

### Grouping datasets on the basis of State

```python
state_basis=df.groupby(by='State').sum()
state_basis_10=state_basis.sort_values(by=['Student Population'],ascending=False).head(10)
state_basis_10
```

### Presenting top 10 most Student Populated College of America using Pie-Chart

```python
fig = px.pie(state_basis_10, values='Student Population', names=state_basis_10.index, title='Most Population of student in 10 States')
fig.show()
```

### Student Population of differnet State on the basis of Public and Private College

```python
fig = px.sunburst(df, path=['Public/Private', 'State'], values='Student Population')
fig.show()
```

### Acceptance Rate in different college

```python
fig=px.histogram(df,x='Acceptance Rate',y='Name')
fig.show()
```

### Total Annual Cost of top 20 most expensive college

```python
df_20=df.sort_values(by=['Total Annual Cost'],ascending=False).head(20)
fig=px.histogram(df_20,x='Name',y='Total Annual Cost')
fig.show()
```
