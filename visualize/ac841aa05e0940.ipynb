```python
seLiga.columns = ["Timestamp",'Age', "snf", 'where', 'how', 'frequency', 'share', 'if share where']
# sns = see news freqently
seLiga
```

```python
where = seLiga['where'].apply(pd.Series)
where
```

```python
where = where.transpose()
where
```

```python
seLiga['where']
```

```python
whereCount = {}
for line in seLiga['where']:
    # print(line)
    values = line.split(';')
    for font in values:
        if (font in whereCount):
            whereCount[font] += 1
        else:
            whereCount[font] = 1
    # print('values', values)
for key, value in whereCount.items():
    print("%s : %d"%(key, value))
```

```python
whereCount
```

```python
seLiga['if share where'].fillna("empty", inplace=True)
seLiga
```

```python
validKeys = ['whatsapp', "instagram", 'whatsapp', 'twitter']

share = {}
for line in seLiga['if share where']:
    # print(line)
    if (line != 'empty'):
        words = line.split(' ')
        # print(words)
        for word in words:
            if (word.lower() in validKeys):
                if (word.lower() in share):
                    share[word.lower()] += 1
                else:
                    share[word.lower()] = 1
share
```

```python
seLiga = seLiga.drop(columns=['if share where', 'Timestamp'])
seLiga
```

```python
snf_map = {
    "Sim":1,
    "Não":0,
}
seLiga = seLiga.replace({'snf':snf_map})
seLiga['snf']
```

```python
how_map = {
    'Vejo o que aparece': 0,
    'Vou atrás de notícias diariamente': 1,
    'Vejo o que me interessa': 2,
    'Me informo de tudo': 3,
}
seLiga = seLiga.replace({'how':how_map})
seLiga['how']
```

```python
frequency_map = {
    'Todo dia':0,
    'Quando preciso me informar':1,
    'Várias vezes ao dia':2,
    'Algumas vezes na semana':3,
    'O mínimo possível':4,
    'Esporadicamente':5,
}
seLiga = seLiga.replace({'frequency':frequency_map})
seLiga['frequency']
```

```python
share_map = {
    'Sim': 1,
    'Não': 0,
    'As vezes': 2,
}
seLiga = seLiga.replace({'share':share_map})
seLiga['share']
```
