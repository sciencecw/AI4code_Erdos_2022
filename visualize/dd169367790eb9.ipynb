========================================<font size='4'>urban sound</font>======================================

```python
# загрузить набор данных "sound"
sound_set = pd.read_csv("/kaggle/input/urbansound8k/UrbanSound8K.csv")
```

```python
# some sound data
sound_set.head(9)
```

```python
# уникальные классы в наборе данных
sound_unique = {}
for el in set(sound_set.classID):
    sound_unique[el] = list(sound_set.classID).count(el)
sound_unique
```

```python
# передний план / фоновые звуки (2 - задний, 1 - передний)
sound_for_back = {}
for el in set(sound_set.salience):
    sound_for_back[el] = list(sound_set.salience).count(el)
sound_for_back
```

```python
# plot unique classes
px.pie(values=sound_unique.values(), names=sound_unique.keys())
```

```python
# передний план / фоновые звуки (2 - задний, 1 - передний)
px.pie(values=sound_for_back.values(), names=sound_for_back.keys())
```

```python
# создаем пути ко всем файлам
sound_set['path'] = sound_set.apply(lambda el: f"/kaggle/input/urbansound8k/fold{el.fold}/{el.slice_file_name}", axis=1)
```

```python
# звук собаки
dog_bark, dog_bark_freq = lr.load(f"{sound_set.iat[0,8]}")
```

```python
# звук ребенка
chld_pl, chld_pl_freq = lr.load(f"{sound_set.iat[1,8]}")
```

```python
# кондиционер
air_cond, air_cond_freq = lr.load(f"{sound_set.iat[4102,8]}")
```

```python
# plot лай собаки
lrd.waveplot(dog_bark, sr=dog_bark_freq)
```

```python
# plot звук ребенка
lrd.waveplot(chld_pl, sr=chld_pl_freq)
```

```python
# plot кондиционер
lrd.waveplot(air_cond, sr=air_cond_freq)
```

```python
# спектрограмма лая собаки
lrd.specshow(lr.amplitude_to_db(abs(lr.stft(dog_bark))), sr=dog_bark_freq, x_axis="time", y_axis="hz")
plt.title("dog bark spect")
plt.colorbar()
```

```python
# спектрограмма детской игры
lrd.specshow(lr.amplitude_to_db(abs(lr.stft(chld_pl))), sr=chld_pl_freq, x_axis="time", y_axis="hz")
plt.title("child play spect")
plt.colorbar()
```

```python
# спектрограмма кондиционера
lrd.specshow(lr.amplitude_to_db(abs(lr.stft(air_cond))), sr=air_cond_freq, x_axis="time", y_axis="hz")
plt.title("air cond spect")
plt.colorbar()
```

<font size='3'>conclusion (urban sound) this dataset contains .wav files of different sounds. you can represent sound files as array of numbers. also this dataset contains foreground and background sounds as well. dataset has 9 different types of audio fragments (child play, dog bark, air cond, etc)</font>
