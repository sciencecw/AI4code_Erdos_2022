```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
df = pd.read_csv('../input/d/felipefiorini/cub-mdio-brasil/dataset.csv')
```

## New columns

```python
df['period'] = df['year'].astype(str) + '.' + ('0' + df["month"].astype(str)).str[-2:]
```

## Agregate

```python
df_agg = df.groupby(['year','label', 'state'],as_index=False).agg({'value':['mean']})
```

```python
#df['value'] = df['value'].astype(np.float16)

for s in df['state'].unique():
    v_period = df_agg[(df_agg['label']=='global') & (df_agg['state']==s)]['year']
    v_value  = df_agg[(df_agg['label']=='global') & (df_agg['state']==s)]['value']
    
    plt.plot(v_period, v_value)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(f'State: {s}')
    plt.show()
```
