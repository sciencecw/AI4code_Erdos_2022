```python
# Import Libraries
import mxnet as mx
import gluoncv as gcv
import matplotlib.pyplot as plt
from mxnet.gluon import nn
from mxnet.gluon.data.vision import datasets, transforms
from IPython import display
import os
import time
```

## Pre-trained YOLO3_darknet model from Gluon
Website: https://cv.gluon.ai/contents.html#

```python
from gluoncv import model_zoo, data, utils

data_paths = [i for i in (os.path.join('../input/car-object-detection/data/testing_images', f) for f in os.listdir('../input/car-object-detection/data/testing_images')) if os.path.isfile(i)]
models=['yolo3_darknet53_coco']

for y in models:
    print("model Name:::"+str(y))
    for i in data_paths:
        net = model_zoo.get_model('yolo3_darknet53_coco',pretrained=True)
        x, img = data.transforms.presets.ssd.load_test(i, short=512)
        class_IDs, scores, bounding_boxs = net(x)
        
        ax=utils.viz.plot_bbox(img, bounding_boxs[0],scores[0], class_IDs[0], class_names=net.classes)
        plt.show()
```
