# Create a neural network model to train using the feature extracted for natural image prediction

```python
from keras import models
from keras.layers import Dense,Dropout
from keras.utils import to_categorical
```

```python
!pip install livelossplot
```

```python
import livelossplot
plot_losses = livelossplot.PlotLossesKeras()
```

```python
xtrain_n = xtrain/xtrain.max()
xtest_n = xtest/xtest.max()
```

```python
ytrain_h = to_categorical(ytrain)
ytest_h= to_categorical(ytest)
```

```python
#Build neural Network Model
model = models.Sequential()
model.add(Dense(1024 ,activation = 'relu' , input_shape = (xtrain.shape[1],)))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(256,activation = 'relu'))
model.add(Dense(256,activation = 'relu'))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(256,activation = 'relu'))
model.add(Dense(256,activation = 'relu'))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(512,activation = 'relu'))
model.add(Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
from keras import optimizers
sgd = optimizers.SGD(lr=0.1)
model.compile(optimizer = sgd,
             loss = 'categorical_crossentropy',
             metrics = ['Accuracy'])
```

```python
xtrain_n.shape
```

```python
model.fit(xtrain_n,ytrain_h,epochs = 1500)
```

```python
model.evaluate(xtrain_n,ytrain_h)
```

```python
model.evaluate(xtest_n,ytest_h)
```
