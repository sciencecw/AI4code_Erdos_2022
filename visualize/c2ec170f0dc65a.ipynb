#### Merging Vaccine DF to the VAERS DF

As noted above there are 445,358 rows in the Vaccine dataframe and 428, 486 rows in the VAERS dataframe
- This will take everything that is in the Vaccine dataframe and connects it to the data in VAERS by VAERS_ID

```python
vaccine_adverse = pd.merge(vaccine, adverse, on = "VAERS_ID")
vaccine_adverse.shape  
```

```python
vaccine_adverse.loc[vaccine_adverse['VAERS_ID'] == 916600]
```

#### Now merging the last dataset, symptoms, to all those VAERS_IDs which have a recorded COVID19 vaccine record. 

Taking the previously merged DF and merging it to symptoms via VAERS_ID

There are 575 022 rows in symptoms, and only those records which are related to the VAERS_IDs will be linked

```python
vaccine_adverse_symptoms = pd.merge(vaccine_adverse, symptoms, on = "VAERS_ID")
vaccine_adverse_symptoms.shape 
```

```python
duplicated_df = vaccine_adverse_symptoms[vaccine_adverse_symptoms.duplicated()]
duplicated_df
```

```python
final_data= vaccine_adverse_symptoms.drop_duplicates()
#By default, it removes duplicate rows based on all columns.
```

```python
final_data.shape
```

```python
covidfinal_data = final_data[final_data.VAX_TYPE == 'COVID19']
covidfinal_data.shape
```

```python
covidfinal_data.head
```

```python
final_file = "2021Covid19VAERSDatasetv2.csv"
print("Saving file: '{}'".format(final_file))
covidfinal_data.to_csv(final_file, index=False, encoding='utf-8')
print("File Saved...")
```
