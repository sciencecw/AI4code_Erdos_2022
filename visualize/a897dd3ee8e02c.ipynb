## Data Transformation and Feature engineering

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1['Gender'] = le.fit_transform(df1['Gender'])
```

```python
df1['MaritalStatus'] = le.fit_transform(df1['MaritalStatus'])
```

```python
df1['PartTimeFullTime']=le.fit_transform(df1['PartTimeFullTime'])     
```

```python
df1.head()
```

```python
df1.columns
```

```python
feature_cols=['Age', 'Gender', 'MaritalStatus','DependentChildren','WeeklyWages', 'PartTimeFullTime','InitialIncurredCalimsCost']
predictor=df1[feature_cols]
response=df1['UltimateIncurredClaimCost']
```

```python
predictor
```

```python
response
```

```python
print(predictor.shape)
print(response.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(predictor, response, test_size = 0.3, random_state =30)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# standard scaling
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

## Model Selection

Here I have selected LinearRegression model to predict Workers Compensation claims using the realistic data give data 
I have selected this model because linear regression works best for predicting the unknown values with least error.

```python
# creating the model
LRM= LinearRegression()
```

```python
LRM.fit(x_train, y_train)
```

```python
print("The Train Score Is :",LRM.score(x_train,y_train))
```

```python
print("The Test Score Is :",LRM.score(x_train,y_train))
```

```python
y_pred =LRM.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
print("The r2 score is :", r2)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

Since the train score and the test score value is very less and r2 value deviates too much LinearRegression model doest work best for the given dataset.
So I have changed the model to RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
# Predicting a new result
y_pred = regressor.predict(x_test)
regressor.score(x_train,y_train)
```

```python
model=regressor.fit(x_train, y_train)
```

```python
model.feature_importances_
```
