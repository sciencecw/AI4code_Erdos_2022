```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

**Importing Datasets**

```python
from sklearn.datasets import load_iris
%matplotlib inline
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

**Dropping Two Features - Sepal Length & Sepal Width**

```python
df.drop(['sepal length (cm)', 'sepal width (cm)'], axis='columns', inplace=True)
df.head()
```

```python
classifier = KMeans(n_clusters=3)
clusters = classifier.fit_predict(df)
clusters
```

```python
df['predicted cluster'] = clusters
df.head()
```

**Scaling Data**

```python
scaler = MinMaxScaler()
scaler.fit(df[['petal length (cm)']])
df['petal length (cm)'] = scaler.transform(df[['petal length (cm)']])
scaler.fit(df[['petal width (cm)']])
df['petal width (cm)'] = scaler.transform(df[['petal width (cm)']])
df.head()
```

```python
df['predicted cluster'].unique()
```

```python
df1 = df[df['predicted cluster']==0]
df2 = df[df['predicted cluster']==1]
df3 = df[df['predicted cluster']==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='orange')
```

**Drawing Elbow Plot for an Optimal Value of K**

```python
sum_squared_errors = []
k_range = range(1,10)
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sum_squared_errors.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared errors')
plt.plot(k_range,sum_squared_errors)
```

***Solution: The point at which the elbow shape is created is 3, that is, our K value or an optimal number of clusters is 3.***
